````python
import os
import pathlib
import random

large_data_base_dir = config["config"]["large_data_path"]

if not os.path.exists("data"):
    os.mkdir("data")

def setup_environment(base_dir: str = large_data_base_dir,
                      depth: int = 5,
                      files_per_dir: int = 700) -> pathlib.Path:
    """
    generator와 list 비교를 위한 깊은 폴더 구조와 파일을 생성합니다.
    """
    base_path = pathlib.Path(base_dir)
    # 기존에 있다면 삭제하고 새로 만듭니다.
    if base_path.exists():
        import shutil
        shutil.rmtree(base_path)
    base_path.mkdir(exist_ok=True)
    print(f"데이터 환경 생성 시작: '{base_dir}'")

    for i in range(files_per_dir * depth):
        print(f"[{str(int(i*100/(files_per_dir * depth))).zfill(2)}%] {i} / {files_per_dir * depth} ", end="\r")
        # 깊은 폴더 구조 생성
        current_dir = base_path
        for d in range(depth):
            current_dir /= f"depth_{d}"
            current_dir.mkdir(exist_ok=True)

        # 각 깊이에 무작위로 파일 생성
        file_count = random.randint(1, files_per_dir)
        for j in range(file_count):
            file_name = f"file_{i}_{j}.txt"
            (current_dir / file_name).touch() # 빈 파일 생성

    print(f"총 {depth}단계 깊이, 최소 {files_per_dir}개의 폴더와 수백 개의 파일 생성 완료.")
    return base_path

setup_environment()
````

## glob  

````python
from glob import glob
from pathlib import Path
import time

start = time.time()

large_data_base_dir = config["config"]["large_data_path"]

data_list = glob(large_data_base_dir + "/**/**/**/**/**/*")
print(f"glob data type : {type(data_list)}\n")

a = data_list.pop()
b = data_list.pop()
print(f"a : {a}")
print(f"b : {b}")

print(f"time : {time.time() - start}")
````

## rglob

````python
from glob import glob
from pathlib import Path
import time

start = time.time()

large_data_base_dir = config["config"]["large_data_path"]

data_gen = Path(large_data_base_dir).rglob("*")
print(f"rglob data type : {type(data_gen)}\n")

a = data_gen.__next__()
b = data_gen.__next__()
print(f"a : {a}")
print(f"b : {b}")

print(f"time : {time.time() - start}")
````
