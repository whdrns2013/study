```python
corpus = [
 ["고양이", "귀엽다", "고양이", "장난", "치다", "고양이", "햇빛", "좋아하다"],
 ["고양이", "종종", "창가", "앉다", "햇빛", "쬐다"],
 ["고양이", "조용한", "공간", "혼자", "쉬다", "좋아하다"],
 ["오늘", "비", "오다", "우산", "들다", "출근"],
 ["새로운", "스마트폰", "공개", "사람들", "새로운", "기능", "기대"],
]
```

```python
word_set = set([y for x in corpus for y in x])
```

```python
word_count = dict()
for word in word_set:
    for doc in corpus:
        if word in doc:
            if word not in word_count.keys():
                word_count[word] = 1
            else:
                word_count[word] += 1
```
