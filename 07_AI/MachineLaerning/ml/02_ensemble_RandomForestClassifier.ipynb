```python
from sklearn.ensemble import RandomForestClassifier
import sklearn.datasets as datasets
import sklearn.model_selection as model_selection
import sklearn.metrics as metrics
import pandas as pd
import pickle
```

```python
# 모델 선언
model = RandomForestClassifier(random_state=11)
```

```python
# 데이터셋 살펴보기
iris = datasets.load_iris()
iris_df = datasets.load_iris(as_frame=True)
iris_target = iris.target
iris_target_name = iris.target_names

print(iris_df)
print(iris_target_name)
print(iris_target)
```

```python
# 데이터셋 스플릿
X_train, X_test, y_train, y_test = model_selection.train_test_split(iris.data,
                                                                    iris.target,
                                                                    test_size=0.2,
                                                                    random_state=11)

print("train 셋 : ", len(X_train), len(y_train), "\n", "test 셋 : ", len(X_test), len(y_test))
```

```python
# 튜닝
```

```python
# 학습
model.fit(X_train, y_train)
```

```python
# 평가
pred = model.predict(X_test)
eval_df = pd.DataFrame(data = zip(pred, y_test), columns=["예측", "정답"])

list_correct = []

for i in range(len(eval_df)):
    # print(eval_df.iloc[i]["예측"])
    list_correct.append(1 if eval_df.iloc[i]["예측"] == eval_df.iloc[i]["정답"] else 0)
eval_df["정답여부"] = list_correct

print("정답률 : " + str(metrics.accuracy_score(y_test, pred)))
eval_df
```
